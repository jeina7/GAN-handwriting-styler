```python
from PIL import Image, ImageOps
import numpy as np
import pandas as pd
import io, os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 72192

def get_data(N):
    file_path = "./tensorflow-hangul-recognition/image-data/hangul-images/"
    hanguls = []
    for i in range(1, N+1):
        with Image.open(file_path + "hangul_{}.jpeg".format(i)) as image:
            hangul = np.array(image)
            hangul = hangul / 255
        hanguls.append(hangul)
    return np.array(hanguls)
```

```python
%%time
hanguls = get_data(N)
print(len(hanguls))
```

```python
hanguls[0].shape
```

```python
plt.figure(figsize=(10, 5))
for i in range(18):
    plt.subplot(3, 6, i+1)
    plt.imshow(hanguls[i], cmap="gray")
    plt.grid()
    plt.axis("off")
plt.show()
```

## reshape image 3D -> 2D

```python
hanguls_reshape = np.array(list(map(lambda i: i.reshape(-1), hanguls)))
hanguls_reshape.shape
```

```python
pd.DataFrame(hanguls_reshape).to_csv("ohe-label-string.csv", index=False)
```

## labels

```python
target_df = pd.read_csv("./tensorflow-hangul-recognition/image-data/labels-map.csv", encoding='utf-8')
print(target_df.shape)
target_df.head()
```

```python
with io.open('./tensorflow-hangul-recognition/labels/256-common-hangul.txt', 'r', encoding='utf-8') as f:
    labels = f.read().splitlines()

len(labels)
```

```python
for i in range(len(labels)):
    ohe_label = [0] * 256
    ohe_label[i] = 1
    target_df['feature'] = list(hanguls_reshape)
    target_df.loc[target_df['label'] == labels[i], 'int_label'] = i
    target_df.loc[target_df['label'] == labels[i], 'ohe_label'] = str(ohe_label)
target_df['int_label'] = target_df['int_label'].astype(int)
target_df.tail()
```
